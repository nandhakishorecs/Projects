Data cleaning

```python
# To remove zero values from the Glucose, Insulin, Blood Pressure, Skin Thickness and BMI as they affect the prediction process
no_zeros=['Glucose','BloodPressure','BMI','Insulin','SkinThickness','BMI']
for column in no_zeros:
    data_diabtes[column]=data_diabtes[column].replace(0,np.NaN)
    mean=int(data_diabtes[column].mean(skipna=True))
    data_diabtes[column]=data_diabtes[column].replace(np.NaN,mean)
```

Independant and dependant variable extraction

```python
x=data_diabtes.iloc[:,0:8]              
y=data_diabtes.iloc[:,8]                
```

Dataset splitted into train and test with test size as 20%

```python
x_train, x_test,y_train, y_test = train_test_split(x,y,random_state=0, test_size=0.2)
```

Feature Scalling

```python
scale=StandardScaler()
x_train=scale.fit_transform(x_train)    
x_test=scale.transform(x_test)          
```

KNN Classifier Defenition

```python
classifier=KNeighborsClassifier(n_neighbors=11,metric='euclidean')
classifier.fit(x_train,y_train)
y_pred1=classifier.predict(x_test)
y_pred1
```

```python
# KNN Model Evaluation:
M=confusion_matrix(y_test,y_pred1)
print(M)
print(f1_score(y_test,y_pred1))
```

```python
print('Accuracy in KNN Classification is',accuracy_score(y_test,y_pred1)) 
```

Prediction by Decision Tree Classifier

```python
# Decision Tree Classifier Model evaluation:
classification = DecisionTreeClassifier(criterion='entropy',max_depth=3)
classification = classification.fit(x_train,y_train)
y_pred2 = classification.predict(x_test)
#Accuracy score
print('Accuracy in Decision Tree Classifier is:',accuracy_score(y_test,y_pred2))

y_pred2
```
